```python
#finding the city which is less than 10000
filtered_data=building[building['price']>=10000]
city_wise_villas=filtered_data.groupby('city')['price'].apply(list)
print(city_wise_villas)
```

```python
# flitering GALT city
flitering_city=building[(building['city']=='GALT')&(building['sq__ft']>800)]
print(flitering_city[['street','sq__ft','sale_date','city']])
```

```python
# finding cheapest villa in CA
CA_villa=building[(building['state']=='CA') & (building['type']=='villa')]
cheapest_property=CA_villa.loc[CA_villa['price'].idxmin()]
print('the cheapest villa property',cheapest_property)
```

```python
#finding more than two beds 
residential=building[(building['type']=='Residential')&(building['beds']>2)]
print(residential[['city','baths','sq__ft','price','type','beds']])
```

```python
#finding top 5 street using 
fliter_residences=building[(building['sq__ft']>=1450)&
                           (building['beds']>=3)&(building['baths']>=2)&
                           (60000>=building['sq__ft'])&
                           (building['sq__ft']<=120000)]
print(fliter_residences.head(5))
```
