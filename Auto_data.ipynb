```python
import matplotlib.pyplot as plt
```

```python
mean_values=auto[["mpg","cylinders","displacement","weight","acceleration","model_year","origin"]].mean()
print(mean_values)
```

```python
plt.bar(mean_values.index,mean_values)
plt.xlabel('headers')
plt.ylabel('mean value')
plt.title('mean value representation')
plt.show()
```

```python
plt.plot(mean_values.index,mean_values)
plt.xlabel('headers')
plt.ylabel('mean value')
plt.title('mean value representation')
plt.show()
```

```python
plt.pie(mean_values,labels=mean_values.index,autopct='%1.1f%%')
plt.title('mean value representation')
plt.show()
```

```python
sns.pairplot(data=auto,vars=['mpg','weight'],height=4,hue='origin')

plt.show()
```

```python
auto['horsepower']=pd.to_numeric(auto['horsepower'],errors='coerce')
result=auto[(auto["acceleration"]>15) & (auto["horsepower"]<170)]['name']
print(result)
```
